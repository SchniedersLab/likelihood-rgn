```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(15,10))
plt.xlabel("Iteration")
plt.ylabel("Training Loss")
plt.title("Training Loss vs Iteration")

losses = []
with open("/Dedicated/schnieders/mrtollefson/machineLearning/training/train8/deterministic-tol-nb/RGN7/runs/CASP7/ProteinNet7Thinning90/1/error.log", "r") as file:
    for line in file:
        loss = float(line.split("\t")[1].split(":")[1])
        losses.append(loss)
plt.plot(np.arange(len(losses)), losses, 'o', label="tol-nb")

losses = []
with open("/Dedicated/schnieders/mrtollefson/machineLearning/training/train8/deterministic-alq-nb/RGN7/runs/CASP7/ProteinNet7Thinning90/1/error.log", "r") as file:
    for line in file:
        loss = float(line.split("\t")[1].split(":")[1])
        losses.append(loss)
plt.plot(np.arange(len(losses)), losses, 'o', label="alq-nb")

losses = []
with open("/Dedicated/schnieders/mrtollefson/machineLearning/training/train8/deterministic-alq-b/RGN7/runs/CASP7/ProteinNet7Thinning90/1/error.log", "r") as file:
    for line in file:
        loss = float(line.split("\t")[1].split(":")[1])
        losses.append(loss)
plt.plot(np.arange(len(losses)), losses, 'o', label="alq-b")

plt.legend()

plt.show()
```
